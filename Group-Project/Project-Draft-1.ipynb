```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('2012-Electricity-Consumption.csv')

print("Data Technical Aspects: ")
print(df.info(), '\n')

print("Buildings using electricity (1 = yes, 2 = no)")
print(df['ELUSED'].value_counts(), '\n')

print("Buildings using natural (1 = yes, 2 = no)")
print(df['NGUSED'].value_counts(), '\n')

print("Buildings using solar (1 = yes, 2 = no)")
print(df['SOUSED'].value_counts(), '\n')

print("Total hours open per week (0-168)")
print(df['WKHRS'].describe(), '\n')
```

What each column means is: $25\%$ is lower than `value_displayed`, $50\%$ is lower than `value_displayed`, and so on.   

```python
print('Number of employees (0 - 6,500)')
print(df['NWKER'].describe())
df['NWKER'].hist(bins = 50)
plt.show
```

```python
pd.set_option('display.float_format', lambda x: '%.f' % x)
print('Square Footage (1,001 - 2,100,000)')
print(df['SQFT'].describe())
df['SQFT'].hist(bins = 50)
plt.show
```

```python
print('Percent Lit when open')
print(df['LTOHRP'].describe())
df['LTOHRP'].hist(bins = 50)
plt.show
```
